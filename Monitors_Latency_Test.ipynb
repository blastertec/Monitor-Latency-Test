
First we calculate mean values of our monitor latency test results



```
import matplotlib.pyplot as plt

DELL_monitor = (225 + 224 + 212) / 3
print("Dell: " + str(round(DELL_monitor)))

PREDATOR_monitor = (196 + 203 + 187) / 3
print("Predator: " + str(round(PREDATOR_monitor)))

DELL_monitor_60 = 218
PREDATOR_monitor_60 = 242
```


Now lets calculate the difference and plot the results




```
difference = DELL_monitor - PREDATOR_monitor
print("Difference is " + str(round(difference)) + " milliseconds")

fig = plt.figure(figsize=(5, 6))
ax = fig.add_subplot(1, 1, 1)

ax.text(1, 150, int(DELL_monitor), fontweight='bold')
ax.text(2, 150, int(PREDATOR_monitor), fontweight='bold')

ax.bar(x=[1, 2],
       height=[DELL_monitor, PREDATOR_monitor],
       color=["Red", "Green"],
       tick_label=["Dell, 60 Hz", "Predator, 240 Hz"],
       width=[0.9],
       linewidth=[1])

plt.title("Human reaction tested on monitors")
plt.ylabel("Latency, milliseconds")
plt.show()
```

```
fig = plt.figure(figsize=(5, 6))
ax = fig.add_subplot(1, 1, 1)

ax.text(1, 150, int(DELL_monitor_60), fontweight='bold')
ax.text(2, 150, int(PREDATOR_monitor_60), fontweight='bold')

ax.bar(x=[1, 2],
       height=[DELL_monitor_60, PREDATOR_monitor_60],
       color=["Red", "Green"],
       tick_label=["Dell, 60 Hz", "Predator, 60 Hz"],
       width=[0.9],
       linewidth=[1])

plt.title("Human reaction tested on monitors")
plt.ylabel("Latency, milliseconds")
plt.show()
```
